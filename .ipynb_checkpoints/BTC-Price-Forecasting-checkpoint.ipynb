```python
## changing column names to 'y' and 'ds' for Fbprophet to identify
f_df['y'] = f_df['4a. close (USD)']
f_df['ds'] = f_df['datetime']
```

```python
df = f_df[['y', 'ds']]
df 
```

```python
#!conda install -c conda-forge fbprophet -y
```

```python
## importing prophet
from fbprophet import Prophet
```

```python
## instantiating model and fiting it to the dataset(df)
m = Prophet(interval_width=0.95, daily_seasonality=True)
model = m.fit(df)
```

```python
## forcasting 100 periods into the future
future = m.make_future_dataframe(periods=100,freq='D')
forecast = m.predict(future)
forecast.head()
```

```python
## ploting the components of the forcast
plot = m.plot_components(forecast)
```

```python
## btc forecast 
plot = m.plot(forecast)
```

```python
forecast.columns
```

```python
## value of predictions from 1000 to 1100
forecast_preds = forecast[['ds','yhat_lower', 'yhat_upper', 'yhat', ]][1000:]
```

```python
fig, ax = plt.subplots(figsize=(16, 4))
x = forecast_preds['ds']
y =  forecast_preds['yhat_lower']
y4 = forecast_preds['yhat_upper']
y2 = forecast_preds['yhat']

ax.plot(x, y, label='yhat_lower')
ax.plot(x, y2, label='yhat')
ax.plot(x, y4, label='yhat_upper')
ax.legend()
plt.show()
```

```python
## ploting the components of the forcast for the next 100 days
plot = m.plot_components(forecast[1000:])
```

```python
forecast_next_5_days_preds = forecast[['ds','yhat_lower', 'yhat_upper', 'yhat', ]][1000:1005]
```

```python
fig, ax = plt.subplots(figsize=(16, 4))
x = forecast_next_5_days_preds['ds']
y =  forecast_next_5_days_preds['yhat_lower']
y4 = forecast_next_5_days_preds['yhat_upper']
y2 = forecast_next_5_days_preds['yhat']

ax.plot(x, y, label='yhat_lower')
ax.plot(x, y2, label='yhat')
ax.plot(x, y4, label='yhat_upper')
ax.legend()
plt.show()
```
